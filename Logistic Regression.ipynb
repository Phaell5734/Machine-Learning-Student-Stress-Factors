```python
import pandas as pd
```

```python
df = pd.read_csv("Student Stress Factors (2).csv")
df.dropna(inplace=True)
df.info()
```

```python
X = df.drop(['How would you rate your stress levels?'], axis=1)
y = df['How would you rate your stress levels?']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 21)
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state = 0).fit(X_train_scaled, y_train)
log_reg.predict(X_train_scaled)
```

```python
log_reg.score(X_train_scaled, y_train)
```

```python
log_reg.score(X_test_scaled, y_test)
```

```python
log_reg1 = LogisticRegression(random_state = 0,
C = 1,
fit_intercept = True,
).fit(X_train_scaled, y_train)
```

```python
log_reg1.score(X_train_scaled, y_train)
```

```python
log_reg1.score(X_test_scaled, y_test)
```
